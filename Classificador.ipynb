```python
#Distancia Euclidiana
def geraMatrizEuclids(x, y):
    #Trasforma para um vetor numpy
    x = np.array(x)
    y = np.array(y)
    
    #Pega o tamanho dos vetores sem os NaN Values
    tamanho_x = len(x[np.logical_not(np.isnan(x))])
    tamanho_y = len(y[np.logical_not(np.isnan(y))])
    
    #Cria uma matriz de zeros com o tamanho dos vetores x e y
    matrizEuclids = np.zeros((tamanho_x, tamanho_y))
        
    #Percorre a matriz adicionando os valores das distancias euclidianas para todos os pontos.
    for i in range(tamanho_x):
        for j in range(tamanho_y):
            matrizEuclids[i][j] = abs(x[0,i] - y[0,j])
    
    return matrizEuclids
#Matriz gerada a partir do ponto h, como é o msm ponto, a distancia da diagonal é zerada.
geraMatrizEuclids(h,h)
```

```python
#Biblioteca pronta de DTW, NÃO É UTILIZADA
from dtw import dtw
def dtw2(x,y):
    
    x = np.array(x).reshape(-1, 1)
    y = np.array(y).reshape(-1, 1)
    manhattan_distance = lambda x, y: np.abs(x - y)

    d, cost_matrix, acc_cost_matrix, path = dtw(x, y, dist=manhattan_distance)
    return acc_cost_matrix[len(x)-1][len(y)-1]
```

```python
#DTW
def dtw1(x,y):
    
    x = np.array(x)
    y = np.array(y)
    
    #Pega o tamanho dos vetores e desconsidera os NaN values
    tam_x = len(x[np.logical_not(np.isnan(x))])
    tam_y = len(y[np.logical_not(np.isnan(y))])
    
    
    
    dtw_matrix = np.zeros((tam_x, tam_y))
            
      
    #Forma a matriz DTW, percorrendo de 1 até o tamanho dos vetores
    for i in range(tam_x):
        for j in range(tam_y):
            
            if i == 0 :
                dtw_matrix[i][j] = abs(y[0][i]-x[0][j]) + dtw_matrix[i][j - 1]
            elif j == 0:
                dtw_matrix[i][j] = abs(y[0][i]-x[0][j]) + dtw_matrix[i - 1][j]
            else:
                dtw_matrix[i][j] = abs(y[0][i]-x[0][j]) + min(dtw_matrix[i-1][j-1], 
                                                               dtw_matrix[i][j-1], 
                                                               dtw_matrix[i-1][j])
            
            
        
    
    return dtw_matrix[tam_x -1][tam_y -1]
```

```python
#Distância Euclidiana

def distanciaEuclidiana(matrix):
    
    tam_x = matrix.shape[0]
    tam_y = matrix.shape[1]
    
    
    
    menor_dimensao = min(tam_x, tam_y)
    
    
    distancia_diagonal = 0
    
    
    # Soma a diagonal primária da matriz euclidiana e divide pela quantidade de pontos do menor vetor
    for i in range(menor_dimensao):
        distancia_diagonal = matrix[i][i] + distancia_diagonal
    distancia_diagonal = distancia_diagonal / menor_dimensao    
    return distancia_diagonal    
```

```python
#Separa o banco de dados por 5: 4 para treino e 1 para teste

#Biblioteca para fazer uma cópia real do dicionário
import copy

def reparte_dados(j):
    
    #Copio todo o dicionario completo para a variavel de treino
    treino = copy.deepcopy(dicionario_por_movimento) 
    
   
    #Crio um dicionário vazio
    teste = {}
    for i in range(12):
        
        #Adiciono apenas o quinto que eu quero ao dicionário de teste
        teste[i+1] = dicionario_por_movimento[i+1][ j*20 : 20 + (j*20) ]   
        
        #Coloco Nan values nas posições que foram adicionadas na variavel de teste
        #Na hora de extrair as distancias, os NaN values são desconsiderados
        #Fiz dessa forma por não conseguir remover esse trecho no dicionário
        
        
        
        treino[i+1][j*20 : 20 + (j*20)] = None
        
        #del treino[i+1][ j*20 : 20 + (j*20)]
        
    

    
    
    
    return teste, treino

teste, treino = reparte_dados(4)
treino
```

```python
#Função de verificação de acerto, se acertou retorna 1, se não retorna 0
def acuracia(vetor, target):
    maior_frequencia = 0
    movimento_mais_frequente = vetor[0][0]
    
    #Percorre todo o vetor
    for i in range(len(vetor)):
        frequencia = 0
        
        #Percorre o vetor verificando quantas vezes a posição i aparece.
        for j in range(len(vetor)):
            
            if vetor[i][0] == vetor[j][0]:
                
                frequencia = frequencia + 1
        #Salva qual é o movimento mais frequente, como o vetor já está ordenado, 
        #em caso de empate, será escolhido sempre o movimento com a menor distancia
        if frequencia > maior_frequencia:
            maior_frequencia = frequencia
            movimento_mais_frequente = vetor[i][0]
    #Verifica se o classificador acertou e dá o valor de retorno correspondente        
    if movimento_mais_frequente == target:
        acuracia = 1
    else: acuracia = 0    
    #print('movimento_mais_frequente: {}'.format(movimento_mais_frequente))    
        
    
    return acuracia
vet = [[1,400],[2,500],[8,550],[1,600],[2,600]]
acuracia(vet, 1)
```

```python
#Classificador KNN


def classificador(k_vizinhos):
     
    acertos_euclids = 0
    acertos_dtw = 0
    #For iterando 5 vezes, para realizar a validação cruzada de 5 passos
    for i in range(5):
        
        
        
        #Muda teste e treino para cada validação
        teste, treino = reparte_dados(i)
        
        print('EPOCH_Verificação:{}'.format(i))
        
        for movimento in range(12):
            for test in range(20):
                #Declaração de um vetor de tuplas de tamanho k_vizinhos,
                #onde o primeiro parametro é o tipo de movimento e o segundo a distancia
                #Zera e inicializa o vetor de resultados a cada teste
                resultado_dtw = []
                resultado_euclids = []
                
                for r in range(k_vizinhos):
                    resultado_dtw.append([None,math.inf]) 
                    resultado_euclids.append([None,math.inf]) 
                    
                    
                print('\033[1m' + 'EPOCH_TEST:mov {}/test {}'.format(movimento+1, test) + '\033[0m')
                
                
                for movimento_treino in range(12):
                    for train in range(100):
                        
                        #Pula a sequencia de NaN Values se o 1º elemento for NaN
                        verifica_vazio = math.isnan(treino[movimento_treino + 1][train : train + 1][0])
                        if verifica_vazio == True:
                            train = train + 21
                            if train > 99:
                                continue

                        linha_teste = teste[movimento + 1][test : test+1]
                        linha_treino = treino[movimento_treino + 1][train : train + 1]

                        #Gera a matriz euclidiana para cálculos
                        matriz = geraMatrizEuclids(linha_teste, linha_treino)
                        
                        #Encontra distancias
                        euclids_dist = distanciaEuclidiana(matriz)
                        dtw_dist = dtw1(linha_teste, linha_treino)

                        #Compara a distancia encontrada na iteração com a maior distacia do vetor ordenado
                        if euclids_dist < resultado_euclids[k_vizinhos-1][1]:

                            #Guarda o vizinho euclids no vetor dos mais próximos
                            resultado_euclids[k_vizinhos - 1] = [movimento_treino + 1, euclids_dist]
                            #Ordena o vetor resultado, tendo como base o 2º parametro da tupla(distancia)
                            resultado_euclids.sort(key=lambda x:x[1])

                        #Compara a distancia encontrada na iteração com a maior distacia do vetor ordenado    
                        if dtw_dist < resultado_dtw[k_vizinhos-1][1]:

                            #Guarda o vizinho tdw no vetor dos mais próximos
                            resultado_dtw[k_vizinhos - 1] = [movimento_treino + 1, dtw_dist]
                            #Ordena o vetor resultado, tendo como base o 2º parametro da tupla(distancia)
                            resultado_dtw.sort(key=lambda x:x[1]) 
                            
                #Salva o movimento real do teste            
                target_test = movimento + 1 
                
                #Soma de acertos de todos os testes
                acertos_euclids = acertos_euclids + acuracia(resultado_euclids, target_test)
                acertos_dtw = acertos_dtw + acuracia(resultado_dtw, target_test)
                
                #Porcentagem calculada por teste executado
                porcent_euclids = (acertos_euclids*100/((movimento*20)+test+1)) / (i+1)
                porcent_dtw = (acertos_dtw*100/((movimento*20)+test+1))/ (i+1)
                
                #Print acertos/acuracia por teste
                print("Acertos: Euclid: {} ({}%) / DTW: {} ({}%)".format(acertos_euclids, porcent_euclids,
                                                                     acertos_dtw, porcent_dtw))
        #Calcula a acuracia média dos testes        
        acuracia_media_dtw = (acertos_dtw / 20) * 100 / (12*20)
        acuracia_media_euclids = (acertos_euclids / 20) * 100 / (12*20)
        
    return acuracia_media_dtw, acuracia_media_euclids
hit_ratio_dtw, hit_ratio_euclids = classificador(10);  
print ('\n\nAcurácia Geral DTW: {}\nAcurácia Geral Euclidiano: {}'.format(hit_ratio_dtw, hit_ratio_euclids))
```
